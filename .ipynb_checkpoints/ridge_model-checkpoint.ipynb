```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_hdb = pd.read_csv("data.csv")
```

```python
df_hdb.head()
```

```python
# Check dtype of features
df_hdb.info() 
```

```python
# Check missing values of features
df_hdb.isnull().sum()
```

```python
# Convert 'remaining_lease' in string to 'remaining_lease_months' in numerical
```

```python
# Function to convert string to months
def convert_to_months(duration_str):
    split_parts = duration_str.split(' ')
    
    if len(split_parts) == 4:
        years = int(split_parts[0])
        months = int(split_parts[2])
        total_months = years * 12 + months
    
    else:
        years = int(split_parts[0])
        total_months = years * 12
        
    return total_months

# Apply the conversion function to the Duration column
df_hdb['remaining_lease_months']  = df_hdb['remaining_lease'].apply(convert_to_months)
```

```python
#drop remaining_lease
df_hdb.drop('remaining_lease', axis=1, inplace=True)
```

```python
df_hdb.head()
```

```python
# convert 'month' in datetime to 'year' and 'month' in numeric
df_hdb['month'] = pd.to_datetime(df_hdb['month'])
df_hdb['year_transac'] = df_hdb['month'].dt.year
df_hdb['month_transac'] = df_hdb['month'].dt.month
```

```python
# drop 'month'
df_hdb.drop(labels='month', axis=1, inplace=True)
```

```python
df_hdb.head()
```

```python
# Shifting columns
```

```python
# Get the columns to be shifted
columns_to_shift_to_left = ['year_transac', 'month_transac', 'remaining_lease_months']

# Get the columns to keep on the left
columns_to_keep = [col for col in df_hdb.columns if col not in columns_to_shift_to_left]

# Reorder the columns by concatenating them in the desired order
shifted_columns = columns_to_shift_to_left + columns_to_keep

# Create a new dataframe with the shifted columns
shifted_df = df_hdb[shifted_columns]

# Overwrite original df
df_hdb = shifted_df
```

```python
df_hdb.head()
```

```python
#EDA
```

```python
df_hdb.info()
```

```python
df_hdb.describe()
```

```python
# split into features x and response y
```

```python
df_hdb_x = df_hdb.drop("resale_price", axis=1)
df_hdb_y = df_hdb["resale_price"]
```

```python
df_hdb_x['storey_range'].unique()
```

```python
df_hdb_y.head()
```

```python
# 4. Transform Data
## Encode
## Scale
## Engineer
```

```python
# Encode categorical columns
```

```python
categorical_columns = df_hdb.select_dtypes(include='object').columns.tolist()
categorical_columns
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in categorical_columns:
    df_hdb_x[col] =  encoder.fit_transform(df_hdb_x[col])

df_hdb_x.head()
```

```python
df_hdb_x.info()
```

```python
# Scale all features
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
array_hdb_x_scaled = scaler.fit_transform(df_hdb_x)
```

```python
df_hdb_x_scaled = pd.DataFrame(array_hdb_x_scaled, columns = df_hdb_x.columns)
```

```python
df_hdb_x_scaled.head()
```

```python
# Quick check on correlation between features and response
```

```python
merged_df = df_hdb_x_scaled.join(df_hdb_y)
correlation_matrix = merged_df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 8))

cmap = sns.diverging_palette(220, 10, as_cmap=True)
center = 0
sns.heatmap(correlation_matrix, cmap=cmap, center=center)
```

```python
# remaining_lease_months, storey_range and floor_area have the biggest impact of resale_price.
```

```python
# 5. Modelling Data
```

```python
## Split into train and test sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_hdb_x_scaled, df_hdb_y, test_size = 0.3, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
predicted_train = lr.predict(X_train)
predicted_test = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("train r2_score is:", round(r2_score(y_train, predicted_train), 3))
print("test r2_score is:", round(r2_score(y_test, predicted_test),3))
```

```python
# using ridge regression
```

```python
from sklearn.linear_model import Ridge
```

```python
X_train_truncated = X_train[['remaining_lease_months',
                             'floor_area_sqm',
                             'town',
                             'storey_range']]
```

```python
X_train_truncated = X_train_truncated.reset_index().drop('index', axis=1)
X_train_truncated
```

```python
X_test_truncated = X_test[['remaining_lease_months',
                             'floor_area_sqm',
                             'town',
                             'storey_range']]
X_test_truncated = X_test_truncated.reset_index().drop('index', axis=1)
X_test_truncated
```

```python
ridge_model = Ridge()
ridge_model.fit(X_train_truncated, y_train)
```

```python
y_pred_train = ridge_model.predict(X_train_truncated)
y_pred_test = ridge_model.predict(X_test_truncated)
```
